# Passo 4 - Cálculo de Métricas

Os passos abaixo visam calcular as métricas em redes complexas que possuem relação com a resiliência e a sustentabilidade das cadeias de suprimentos. 

Passo 4.1 *Centralidade combinada*

---> Número de nós vizinhos que o nó princial possui. Quanto menor a centralidade do nó, menor o efeito que ele exerce na rede como um todo, caso ela passe por uma perturbação.

**Contribuição: Resiliência**

```python
#Grau de centralidade do nó
nx.degree_centrality(G1)
```

--> Calcúlo para associação entre o peso do nó e o grau de centralidade do nó. Essas duas métricas combinadas indicam a importância do nó para a rede como um todo. 

**Contribuição: Resiliência**


```python
# Cálculo para associação entre peso do nó e grau de centralidade do nó
node_weights = nx.get_node_attributes(G1,"weight")
node_weights = {k:float(v) for (k,v) in node_weights.items()}
dc = nx.degree_centrality(G1)

new_node_weigh = {}
for node_weight_k, node_weight_v in node_weights.items():
  for d_k, d_v in dc.items():
    if node_weight_k == d_k:
      new_node_weigh.update({node_weight_k: (node_weight_v * d_v)})

new_node_weight_list = [v for k, v in new_node_weigh.items()]

print(new_node_weight_list)
```

```python
# Gráfico visual dos nós que possuem maior importância na rede
import pandas as pd
import matplotlib.pyplot as plt

weights_df = pd.DataFrame(
    {'node_name': [k for k in node_weights.keys()],
     'node_weight': [v for v in node_weights.values()]}
)

dc = nx.degree_centrality(G1)
dc_df = pd.DataFrame(
    {'node_name': [k for k in dc.keys()],
     'node_dc': [v for v in dc.values()]}
)

nodes_df = weights_df.merge(dc_df, 'left', on='node_name')
nodes_df['new_node_weight'] = nodes_df.node_weight * nodes_df.node_dc
nodes_df.sort_values(by=['new_node_weight'], ascending=False, inplace=True)
nodes_df.reset_index(drop=True, inplace=True)

nodes_df
nodes_df.plot(kind='bar', x='node_name', y='new_node_weight', figsize=(20, 10))
```

```python
#Primeiro e segundo nó com maiores impactos caso sofram uma perturbação 
print(f'Nós com maior peso:\n{nodes_df.node_name.iloc[0]} = {nodes_df.new_node_weight.iloc[0]}\n{nodes_df.node_name.iloc[1]} = {nodes_df.new_node_weight.iloc[1]}')
```

Passo 4.2 Comprimento do caminho médio

---> O caminho médio é a média de todas as distâncias dos caminhos que conectam todos os pares de nós possíveis. Quanto menor o caminho médio, mais rápido uma rede se recupera após uma perturbação e mais sustentável é a cadeia de suprimentos do ponto de vista de pegada de carbono. 

**Contribuição: Resiliência e Sustentabilidade**

```python
#comprimento médio do caminho
nx.average_shortest_path_length(G1,weight="value")
```

Passo 4.3 Densidade

--> Uma rede mais densa exige mais custos de operação e tem uma maior dificuldade em manter ações coordenadas. No entanto, há uma menor chance da rede sofrer uma perturbação. Além disso, quanto mais densa a rede, mais sustentável, uma vez que os caminhos percorridos para atendimento de toda a rede são menores. 

**Contribuição: Resiliência e Sustentabilidade**

```python
#densidade da rede
nx.density(G1)
```

Passo 4.4 Complexidade de nós e arestas (escala) 

--> Complexidade de nós e arestas (escala): Conforme a complexidade da rede aumenta, maior a resiliência, uma vez que existem mais opções de fornecimento, por uma questão de conexões.
 
**Contribuição: Resiliência**

```python
#Contagem do número de portos
def count_values_in_dict(input, value):
  value_list = [x for x in input.values() if x == value]
  list_size = len(value_list)
  return list_size

numero_de_portos = count_values_in_dict(nx.get_node_attributes(G1,"categoria1"), 'Porto')
print(numero_de_portos)
```

```python
E = nx.number_of_edges(G1)
N = nx.number_of_nodes(G1)
Complexidade = N/(E+numero_de_portos)
print(Complexidade)
```

Passo 4.5 Coeficiente de Clusterização

-->Mede o grau em que os nós da rede tendem a se agrupar. Esse fator é importante para a análise de problemas logísticos e otimização de rotas, sendo um tema importante para a sustentabilidade. Quanto mais otimizada são as rotas, mais sustentável é o sistema. 

**Contribuição: Sustentabilidade**

```python
#Coeficiente de Clusterização Global
nx.average_clustering(G1,None,"min")
```

```python
#Coeficiente de clusterização de cada um dos nós
nx.clustering(G1,None,"value")
```

# Passo 5 - Resultado Final

O output final abaixo mostra em um gráfico de teia qual a contribuição de cada uma das métricas para o aumento de resiliência e sustentabilidade da rede. Quanto maior o parâmetro, mais forte é a caracteristica de resiliência e/ou sustentabilidade da rede

```python
#Gráfico final - Output para operação
import plotly.graph_objects as go
import plotly.offline as pyo
import networkx as nx


# Definindo o valor máximo do gráfico
PLOT_RANGE_MAX = 0.05

categories = ['Comprimento do caminho (R&S)', 'Densidade (R&S)', 'Complexidade (R)', 'Grau de Clusterização (S)']
categories = [*categories, categories[0]]

cadeia_baseline = [nx.average_shortest_path_length(G1,weight="value")/100,
                nx.density(G1),
                Complexidade/100,
                nx.average_clustering(G1,None,"min")
                ]
cadeia_baseline = [*cadeia_baseline, cadeia_baseline[0]]


fig = go.Figure(
    data=[
        go.Scatterpolar(r=cadeia_baseline, theta=categories, name='Cadeia Teste', fill='toself')
    ],
    layout=go.Layout(
        title=go.layout.Title(text='Sustentabilidade e Resiliência'),
        polar={'radialaxis': {'visible': True, 'range': [0, PLOT_RANGE_MAX]}},
        showlegend=True
    )
)

pyo.plot(fig)
fig.show()
```
